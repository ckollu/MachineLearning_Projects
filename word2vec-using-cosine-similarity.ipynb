```python
import re

def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
movies = []
for x in range(1000) : 
    movies.append(df["Movie Description"][x].strip()+" "+df["Plot"][x].strip())
```

```python
from bs4 import BeautifulSoup
from tqdm import tqdm
pre_processed_movies = []

for sentance in tqdm(movies):
    
    sentance = re.sub(r"http\S+", "", sentance)
    sentance = BeautifulSoup(sentance, 'lxml').get_text()
    sentance = decontracted(sentance)
    sentance = re.sub("\S*\d\S*", "", sentance).strip()
    sentance = re.sub('[^A-Za-z]+', ' ', sentance)
    # https://gist.github.com/sebleier/554280
    sentance = ' '.join(e.lower() for e in sentance.split() if e.lower() not in stopwords.words('english'))
    pre_processed_movies.append(sentance.strip())
```

```python
df["pre_processed_movies"] = pre_processed_movies
```

```python
words=[]
for i in tqdm(pre_processed_movies):
    for j in i.split():
        words.append(j)
print('Number of unique words is : ',len(set(words)))
```

```python
import gensim
import numpy as np
model = gensim.models.KeyedVectors.load_word2vec_format('../input/nlpword2vecembeddingspretrained/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
a = list(model.key_to_index.keys())
embedding_vector = {}
for x in a :
    embedding_vector[x] = model[x]
```

```python
def sent_vectors(sentance,require_preprocessing):
    
    if(require_preprocessing):
        sentance = re.sub(r"http\S+", "", sentance)
        sentance = BeautifulSoup(sentance, 'lxml').get_text()
        sentance = decontracted(sentance)
        sentance = re.sub("\S*\d\S*", "", sentance).strip()
        sentance = re.sub('[^A-Za-z]+', ' ', sentance)
        # https://gist.github.com/sebleier/554280
        sentance = ' '.join(e.lower() for e in sentance.split() if e.lower() not in stopwords.words('english'))

    sent_vec = np.zeros(300) 
    cnt_words =0; 
    for word in sentance.split(): 
        if word in embedding_vector:
            vec = embedding_vector[word]
            sent_vec += vec
            cnt_words += 1
    if cnt_words != 0:
        sent_vec /= cnt_words
    return np.array(sent_vec)
```
