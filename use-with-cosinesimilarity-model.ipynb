```python
import tensorflow as tf
import tensorflow_hub as hub


module_url = "https://tfhub.dev/google/universal-sentence-encoder/4" 
model = hub.load(module_url)
print("USE model %s loaded" % module_url)
def embed(input):
    return model(input)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
inputs=input('enter query :')

vector=embed([inputs])
vector=np.array(vector)
score = []
for i in range(len(word_Embeddings)):
    s = cosine_similarity(vector,np.array(word_Embeddings[i]).reshape(1,512))
    score.append(s[0][0])
    
    
    
```

```python
movie_ids = sorted(range(len(score)), key=lambda i: score[i], reverse=True)[:5]
movie_ids_Scores = [score[x] for x in movie_ids]
```

```python
pred =[]
thres = 0.65
for x in range(len(eva)):
    x1 = int(eva["MOVIE A"][x][1:])
    x2 = int(eva["MOVIE B"][x][1:])
    movie_a = df["Movie Description"][x1].strip()+" "+df["Plot"][x1].strip()
    movie_b = df["Movie Description"][x2].strip()+" "+df["Plot"][x2].strip()
    
    vector1 = embed([movie_a])
    vector2 = embed([movie_b])
    vector = np.array(vector)
    s = cosine_similarity(vector1, vector2)
    
    if s > 0.62:
        pred.append(1)
    else:
        pred.append(0)
eva["Prediction"] = pred
```

```python
from sklearn.metrics import classification_report

target_names = ["Similar", "Not Similar"]
print(classification_report(eva["LABEL"], eva["Prediction"], target_names = target_names))
```
